```python
import pandas
from matplotlib import pyplot
import datetime
import numpy
%matplotlib inline
pandas.options.display.mpl_style = 'default'
```

```python
f = "../data/results-figure2-10.csv"
df = pandas.read_csv(f)
diff = df['Real']-df['Sample1']
diff[diff != 0]
#type(diff.index[0])
```

```python
def parse_date(d):
    v = d.split("/")
    return datetime.date(int(v[2]), int(v[0]), int(v[1]))

df["day"] = df["Date"].apply(parse_date)
df.set_index("day", inplace=True)
```

```python
columns = [col for col in df.columns if "Sample" in col]
columns
```

```python
df['high'] = 0.0
df['low'] = 0.0
for j, row in df[columns].iterrows():
    h = numpy.percentile(row.values, 97.5)
    l = numpy.percentile(row.values, 2.5)
    df.ix[j, 'high'] = h
    df.ix[j, 'low'] = l


```

```python
ix = 2700
ix2 = 1000
pyplot.figure(figsize=(15,8))
pyplot.plot(df.index[-ix:ix+ix2], df[columns][-ix:ix+ix2], alpha=0.1, color='blue')
pyplot.plot(df.index[-ix:ix+ix2], df["high"][-ix:ix+ix2], label="97.5% Level")
pyplot.plot(df.index[-ix:ix+ix2], df["low"][-ix:ix+ix2], label="2.5% Level")
pyplot.plot(df.index[-ix:ix+ix2], df["Real"][-ix:ix+ix2], color="red", label="DJIA")
pyplot.draw()
pyplot.legend(loc=2)
pyplot.xlabel("Date")
pyplot.ylabel("Dow Jones Industrial Average")
pyplot.title("100 Iterations of Generic Prediction\nSampling 10 Times")
pyplot.savefig("/home/tj/Downloads/figure2.svg")
pyplot.savefig("/home/tj/Downloads/figure2.png")

#df[columns][-500:].plot(alpha=0.25, legend=False)
#df[['high', 'low', 'Real']][-500:].plot
```
