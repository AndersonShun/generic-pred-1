```python
import pandas
from matplotlib import pyplot
import datetime
import numpy
%matplotlib inline
pandas.options.display.mpl_style = 'default'
```

```python
size_sample = 1000
f = "../data/global-temp-predict-%i.csv" % size_sample
df = pandas.read_csv(f)
diff = df['Real']-df['SampleSize1']
diff[diff != 0]
type(diff.index[0])
df["Year"] = df.index + 1880
```

```python
columns = [col for col in df.columns if "Sample" in col]
```

```python
df['high'] = 0.0
df['low'] = 0.0
for j, row in df[columns].iterrows():
    h = numpy.percentile(row.values, 97.5)
    l = numpy.percentile(row.values, 2.5)
    df.ix[j, 'high'] = h
    df.ix[j, 'low'] = l


```

```python
ix = 2700
ix2 = 1000
pyplot.figure(figsize=(15,8))
pyplot.plot(df['Year'], df[columns], alpha=0.1, color='blue')
#pyplot.plot(df.index[-ix:ix+ix2], df["high"][-ix:ix+ix2], label="97.5% Level")
#pyplot.plot(df.index[-ix:ix+ix2], df["low"][-ix:ix+ix2], label="2.5% Level")
pyplot.plot(df['Year'], df["Real"], color="red", label="DJIA")
pyplot.draw()
#pyplot.legend(loc=2)
pyplot.xlabel("Year")
pyplot.ylabel("Temperature (Celsius)")
pyplot.title("100 Iterations of Generic Prediction\nSampling %i Times" % size_sample)
pyplot.savefig("../figures/global-temp-%i.svg" % size_sample)
pyplot.savefig("../figures/global-temp-%i.png" % size_sample)

#df[columns][-500:].plot(alpha=0.25, legend=False)
#df[['high', 'low', 'Real']][-500:].plot
```
